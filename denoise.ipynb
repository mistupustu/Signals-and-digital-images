```python
import numpy as np
from scipy import fftpack
import matplotlib.pyplot as plt
```

Nakladanie szumu na obraz

```python
img = plt.imread('dog.jpg').astype(float)
img = img / img.max()

noise = np.random.random(img.shape)
noisy_img = img + noise

plt.imshow(noisy_img)
plt.title('Noisy Image')
plt.show()
img = noisy_img
```

Odszumianie poprzez transformacje fouriera na kolorowym obrazie

```python
img_fft = img.copy().astype(np.complex128)
for channel in range(img_fft.shape[2]):
    img_fft[:, :, channel] = fftpack.fft2(img[:, :, channel])
    plt.imshow(np.log(np.abs(img_fft[:, :, channel])))
    plt.colorbar()
    plt.title(f'FFT in channel number {channel}')
    plt.show()
```

```python
threshold = 0.1
img_filtered = img_fft.copy()
rows, columns, channels_count = img_filtered.shape
for channel in range(img.shape[2]):
    img_filtered[int(rows*threshold):int(rows*(1-threshold)), :, channel] = 0
    img_filtered[:, int(columns*threshold):int(columns*(1-threshold)), channel] = 0
    plt.imshow(np.log(np.abs(img_filtered[:, :, channel])))
    plt.colorbar()
    plt.title(f'Filtered image in channel number {channel}')
    plt.show()
```

```python
img_denoised = img.copy()
for channel in range(img.shape[2]):
    img_denoised[:, :, channel] = fftpack.ifft2(img_filtered[:, :, channel]).real
img_denoised = np.clip(img_denoised, 0, 1)
plt.imshow(img_denoised)
f, ax = plt.subplots(1,2)
ax[0].imshow(img)
ax[0].set_title('Original image')
ax[1].imshow(img_denoised)
ax[1].set_title('Denoised image')

#Zauważyć można, że obraz lekko stracił na jakości, jednak szum jest mniej widoczny.
```

Odszumianie obrazu w odcieni szarości za pomocą transormacji falkowej

```python
import pywt
from skimage import io
```

```python
img_gray = io.imread('img.jpg', as_gray=True)
plt.imshow(img_gray, cmap='gray')
img_gray.shape
```

```python
wavelet = 'sym4'
threshold = 0.1
```

```python
coeffs = pywt.dwt2(img_gray, wavelet)
cA, (cH, cV, cD) = coeffs

cA_thresh = pywt.threshold(cA, threshold)
cH_thresh = pywt.threshold(cH, threshold)
cV_thresh = pywt.threshold(cV, threshold)
cD_thresh = pywt.threshold(cD, threshold)

denoised = pywt.idwt2((cA_thresh, (cH_thresh, cV_thresh, cD_thresh)), wavelet)
```

```python
f, ax = plt.subplots(1,2)
ax[0].imshow(img_gray, cmap='gray')
ax[0].set_title('Original image')
ax[1].imshow(denoised, cmap='gray')
ax[1].set_title('Denoised image')

#Można zauważyć różnicę między odszumionym obrazem a oryginalnym.
```

Odszumianie przez rozmycie gaussowskie

```python
from scipy.ndimage import convolve
def Gauss_blur_5_on_5(image):
  gauss_5_on_5 = np.array([[1, 4, 6, 4, 1], [4, 16, 24, 16, 4], [6, 24, 32, 24, 6], [4, 16, 24, 16, 4], [1, 4, 6, 4, 1]])/256
  result = np.zeros_like(image)
  for channel in range(image.shape[2]):
    result[:,:,channel] = convolve(image[:, :, channel], gauss_5_on_5, mode='constant', cval=0.0)
  return result
```

```python
img_blurred = Gauss_blur_5_on_5(img)
img_blurred = np.clip(img_blurred, 0, 1)
plt.imshow(img_blurred)
```

```python
f, ax = plt.subplots(1,3)
ax[0].imshow(img)
ax[0].set_title('Original image')
ax[1].imshow(img_denoised)
ax[1].set_title('Denoised fourier')
ax[2].imshow(img_blurred)
ax[2].set_title('Denoised gauss blur')
plt.subplots_adjust(wspace=0.3)
plt.show()

#Można zauważyć, że dla tego przypadku metoda rozmycia gaussowskiego poradziła sobie tak samo, a może nawet i lepiej niż metoda fouriera.
#Jednakże najlepsza metoda zależy od zdjęcia i nie zawsze może się to sprawdzić
```
